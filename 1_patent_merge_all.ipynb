```python
import pandas as pd
import os

# Folder path containing the text files
folder_path = 'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data/patent_derwent_wos'

# Output Excel file name
output_file = 'merged_output.xlsx'

# List to store the data frames
dfs = []

# Loop through the files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        # Read the tab-delimited file into a DataFrame
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, delimiter='\t')
        dfs.append(df)

# Merge the data frames into a single DataFrame
merged_df = pd.concat(dfs)

# Save the merged DataFrame to an Excel file
merged_df.to_excel(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/first_clean/patent_all_merge.xlsx')
merged_df.to_pickle(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/first_clean/patent_all_merge.pkl')
merged_df
```
