# Import and concatenate data files of each year in year file

```python
# import OS module
import os
 
# Get the list of all files and directories
#path = "D:\RESEARCH IN AFRICA\DATA\wos_cc_20042022"
path = "D:\PROJECT\YORK UNIVERSITY\BACTERIOPHAGES\data\scholar_wos"
year_list = os.listdir(path)
 
print("Files and directories in '", path, "' :")
 
# prints all files
year_list
```

```python
 # create an Empty DataFrame object
df_all_year_merge = pd.DataFrame()
for year in year_list :
    # create an Empty DataFrame object
    df_year = pd.DataFrame()
    col_to_remove = ['Book Authors','Book Editors','Book Group Authors','Book Author Full Names','Group Authors','Book Series Title','Book Series Subtitle','Conference Title',
                     'Conference Date','Conference Location','Conference Sponsor','Conference Host','Cited References','Part Number','Supplement','Special Issue',
                     'Meeting Abstract','Book DOI','Web of Science Record']
    
    # Get the list of all files and directories in specific year folder
    path_year = f"{path}/{year}"
    year_elt_list = os.listdir(path_year)
    
    for elt in year_elt_list:
        #Import data
        data = pd.read_excel(f"{path_year}/{elt}")
        # drop col_to_remove
        data=data.drop(columns=col_to_remove)
        # add data to Empty DataFrame object created 
        df_year = pd.concat([df_year, data])
    
   # fill nan values in publication year colomns by the year of the main folder
    df_year['Publication Year'].fillna(int(f'{year}'), inplace=True)
    
    df_all_year_merge = pd.concat([df_all_year_merge, df_year])
    
    # uniformized the index
    df_all_year_merge = df_all_year_merge.reset_index().drop('index', axis=1)
        



#print the last
df_all_year_merge
```

```python
df_all_year_merge = df_all_year_merge.applymap(lambda s: s.lower() if type(s) == str else s)

#  remove duplicate row based on Accession_Number + Title + Source_Title 
df_all_year_merge.drop_duplicates(subset= ['Cited Reference Count','Source Title','Article Title'], keep="last", inplace = True)

#save data
df_all_year_merge.to_pickle(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/first_clean/data_scholar_all_year_merge.pkl')
df_all_year_merge.to_csv(f'D:\PROJECT/YORK UNIVERSITY/BACTERIOPHAGES/data_analyse/first_clean/data_scholar_all_year_merge.csv')

df_all_year_merge
```
